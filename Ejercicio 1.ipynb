```python
# Importar las librerías necesarias
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Paso 1: Cargar el archivo CSV
df = pd.read_csv('./Data/trx_tarjetas_credito.csv')
```

```python
# Paso 2: Limpieza de los datos
# Convertir la columna 'Date' al formato de fecha
df['Date'] = pd.to_datetime(df['Date'], format='%d-%b-%y')
```

```python
# Verificar valores faltantes
missing_values = df.isnull().sum()
print("Valores faltantes por columna:\n", missing_values)
```

```python
# Paso 3: Resumen estadístico de la columna 'Amount'
amount_summary = df['Amount'].describe()
print("\nResumen estadístico de los montos de transacción:\n", amount_summary)
```

```python
# Paso 4: Visualización de los datos

# Gráfico 1: Distribución de los montos de transacción
plt.figure(figsize=(10, 6))
sns.histplot(df['Amount'], bins=30, kde=True, color='blue')
plt.title('Distribución de los Montos de Transacción')
plt.xlabel('Monto')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Gráfico 2: Número de transacciones por tipo de tarjeta
plt.figure(figsize=(10, 6))
sns.countplot(x='Card Type', hue='Card Type', data=df, palette='viridis', order=df['Card Type'].value_counts().index, legend=False)
plt.title('Número de Transacciones por Tipo de Tarjeta')
plt.xlabel('Tipo de Tarjeta')
plt.ylabel('Número de Transacciones')
plt.xticks(rotation=45)
plt.show()
```
